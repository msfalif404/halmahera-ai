````python
import json
import os

from elasticsearch import Elasticsearch, helpers
from langchain_ibm import WatsonxEmbeddings
from ibm_watsonx_ai.metanames import EmbedTextParamsMetaNames

from dotenv import load_dotenv
import os

load_dotenv(dotenv_path="../.env", override=True)

ES_INDEX = "scholarship_vector_index"
HOST_ELASTICSEARCH = os.getenv('HOST_ELASTICSEARCH')
API_KEY_ELASTICSEARCH = os.environ.get("API_KEY_ELASTICSEARCH")
WATSONX_API_KEY = os.getenv('WATSONX_API_KEY')
WATSONX_URL = os.getenv('WATSONX_URL')
WATSONX_PROJECT_ID = os.getenv('WATSONX_PROJECT_ID')
WATSONX_MODEL_ID = os.getenv('WATSONX_MODEL_ID')

try:
    es_client = Elasticsearch([HOST_ELASTICSEARCH], api_key=API_KEY_ELASTICSEARCH)
    if not es_client.ping():
        raise ConnectionError("Koneksi ke Elasticsearch gagal.")
    
    print("Berhasil terhubung ke Elasticsearch!")
except Exception as e:
    print(f"Error koneksi: {e}")
    exit()

embed_params = {
    EmbedTextParamsMetaNames.TRUNCATE_INPUT_TOKENS: 512, 
    EmbedTextParamsMetaNames.RETURN_OPTIONS: {"input_text": True},
}

try:
    watsonx_embedding = WatsonxEmbeddings(
        apikey=WATSONX_API_KEY,
        model_id=WATSONX_MODEL_ID,
        url=WATSONX_URL,
        project_id=WATSONX_PROJECT_ID,
        params=embed_params,
    )

    EMBEDDING_DIM = 384 
    print(f"Watsonx Embedding Model '{WATSONX_MODEL_ID}' siap. Dimensi diasumsikan: {EMBEDDING_DIM}")
except Exception as e:
    print(f"Error inisialisasi WatsonxEmbeddings: {e}")
    exit()
    
def delete_index_if_exists(index_name):
    try:
        es = Elasticsearch([HOST_ELASTICSEARCH])

        es.indices.delete(index='*', ignore=[400, 404])
        es.indices.delete(index=index_name, ignore=[404])

        print("‚úÖ Semua indices berhasil dihapus!")
    except Exception as e:
        print(f"Error menghapus indeks: {e}")

def load_data_from_json(filepath="beasiswa.json"):
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            data = json.load(f)
            
        print(f"Berhasil memuat {len(data)} dokumen dari {filepath}.")
        return data
    except FileNotFoundError:
        print(f"Error: File '{filepath}' tidak ditemukan.")
        return []
    except json.JSONDecodeError:
        print(f"Error: File '{filepath}' bukan JSON yang valid.")
        return []

def create_vector_index():
    if es_client.indices.exists(index=ES_INDEX):
        print(f"Indeks '{ES_INDEX}' sudah ada. Melewati pembuatan mapping.")
        return

    mapping = {
        "mappings": {
            "properties": {
                "embedding": {
                    "type": "dense_vector",
                    "dims": EMBEDDING_DIM  
                },
                
                "scholarship_name": {"type": "text"},
                "scholarship_desc": {"type": "text"},
                "scholarship_location": {"type": "keyword"},
                
                "scholarship_deadline": {"type": "text"},
                "scholarship_university": {"type": "text"},
                "scholarship_degree": {"type": "text"},
                "scholarship_field": {"type": "text"},
                "tags": {"type": "text"},
                "scholarship_link": {"type": "text"}
            }
        }
    }
    es_client.indices.create(index=ES_INDEX, ignore=400, body=mapping)
    print(f"Indeks '{ES_INDEX}' berhasil dibuat dengan mapping yang sesuai.")


def generate_and_index_data(documents):
    actions = []
    
    BATCH_SIZE = 10 
    
    for i in range(0, len(documents), BATCH_SIZE):
        batch_docs = documents[i:i + BATCH_SIZE]

        texts_to_embed = [
            f"{doc['scholarship_name']}. {doc['scholarship_desc']}" 
            for doc in batch_docs
        ]
        
        print(f"Menghasilkan embedding untuk batch {i//BATCH_SIZE + 1}...")
        
        try:
            vectors = watsonx_embedding.embed_documents(texts_to_embed)
            print(f"Embedding untuk batch {i//BATCH_SIZE + 1} berhasil dihasilkan.")
        except Exception as e:
            print(f"Gagal mendapatkan embedding untuk batch {i//BATCH_SIZE + 1}: {e}. Melompati batch ini.")
            continue

        for j, doc in enumerate(batch_docs):
            try:
                vector_for_doc = vectors[j]
                if isinstance(vector_for_doc, list) and isinstance(vector_for_doc[0], list):
                    final_vector = vector_for_doc[0] 
                else:
                    final_vector = vector_for_doc
            except IndexError:
                print(f"Peringatan: Tidak ada vektor untuk dokumen {doc['scholarship_name']}")
                continue

            document_to_index = {
                "_index": ES_INDEX,
                "_source": {
                    "scholarship_name": doc["scholarship_name"],
                    "scholarship_desc": doc["scholarship_desc"],
                    "scholarship_location": doc["scholarship_location"],
                    
                    "scholarship_deadline": doc["scholarship_deadline"],
                    "scholarship_university": doc["scholarship_university"],
                    "scholarship_degree": doc['scholarship_degree'],
                    "scholarship_field": doc['scholarship_field'],
                    "tags": doc['tags'],
                    "scholarship_link": doc['scholarship_link'],
                    
                    "embedding": final_vector  
                }
            }
            actions.append(document_to_index)

    print(f"\nMemulai indexing untuk {len(actions)} dokumen...")
    success, errors = helpers.bulk(es_client, actions, raise_on_error=False)
    print(f"Indeks selesai. Berhasil mengindeks {success} dokumen. Total error: {len(errors)}")

def perform_knn_search(query_text, k=5):
    print(f"\nüîç Mencari: '{query_text}'...")
    
    query_vector = watsonx_embedding.embed_query(query_text.lower())
    
    search_body = {
        "size": k,
        "query": {
            "script_score": {
                "query": {"match_all": {}},
                "script": {
                    "source": "cosineSimilarity(params.query_vector, 'embedding') + 1.0",
                    "params": {
                        "query_vector": query_vector
                    }
                }
            }
        },
        "_source": [
            "scholarship_name",
            "scholarship_location",
            "scholarship_desc",
            "scholarship_deadline"
        ]
    }
    
    try:
        results = es_client.search(index=ES_INDEX, body=search_body)
        
        print(f"\n*** Hasil Pencarian untuk: '{query_text}' ***")
        print(f"Found {len(results['hits']['hits'])} results\n")
        
        for i, hit in enumerate(results['hits']['hits'], 1):
            print(f"{i}. Skor: {hit['_score']:.4f}")
            print(f"   Nama: {hit['_source']['scholarship_name']}")
            print(f"   Lokasi: {hit['_source']['scholarship_location']}")
            print(f"   Lokasi: {hit['_source']['scholarship_desc']}")
            if 'scholarship_deadline' in hit['_source']:
                print(f"   Deadline: {hit['_source']['scholarship_deadline']}")
            print()
            
        return results
        
    except Exception as e:
        print(f"‚ùå Error: {e}")
        import traceback
        traceback.print_exc()
        return None
    
if __name__ == "__main__":
    scholarship_data = load_data_from_json("beasiswa.json")
    regerate_index = False

    if regerate_index:
        delete_index_if_exists(ES_INDEX)
        create_vector_index()
        generate_and_index_data(scholarship_data)

    if scholarship_data:
        # perform_knn_search("scholarship for Master studies in Sweden")
        # perform_knn_search("beasiswa S2 di Eropa")
        # perform_knn_search("PhD funding in Computer Science")
        perform_knn_search("Mau cari beasiswa untuk orang disabilitas dong")
````

````python
import requests
import json

BASE_URL = "http://localhost:8000"

def test_get_all_scholarships():
    """Test endpoint untuk mengambil semua beasiswa"""
    print("üß™ Testing GET /")
    
    response = requests.get(f"{BASE_URL}/", params={"limit": 10})
    
    print(f"Status Code: {response.status_code}")
    if response.status_code == 200:
        data = response.json()
        print(f"Count: {data.get('count', 0)}")
        print(f"Results: {len(data.get('results', []))} items")
        print("‚úÖ Test passed")
    else:
        print(f"‚ùå Test failed: {response.text}")
    print("-" * 50)

def test_search_scholarships():
    """Test endpoint untuk pencarian beasiswa"""
    print("üß™ Testing GET /search")
    
    test_queries = [
        "computer science",
        "engineering scholarship",
        "PhD program"
    ]
    
    for query in test_queries:
        print(f"Query: '{query}'")
        response = requests.get(f"{BASE_URL}/search", params={"query": query, "k": 5})
        
        print(f"Status Code: {response.status_code}")
        if response.status_code == 200:
            data = response.json()
            print(f"Query: {data.get('query')}")
            print(f"Results: {len(data.get('results', []))} items")
            print("‚úÖ Test passed")
        else:
            print(f"‚ùå Test failed: {response.text}")
        print("-" * 30)

def test_api_health():
    """Test koneksi ke API"""
    print("üß™ Testing API Health")
    
    try:
        response = requests.get(f"{BASE_URL}/", timeout=5)
        if response.status_code == 200:
            print("‚úÖ API is running")
        else:
            print(f"‚ö†Ô∏è API returned status: {response.status_code}")
    except requests.exceptions.ConnectionError:
        print("‚ùå Cannot connect to API. Make sure server is running.")
    except requests.exceptions.Timeout:
        print("‚ùå API request timeout")
    print("-" * 50)

if __name__ == "__main__":
    print("üöÄ Starting API Tests")
    print("=" * 50)
    
    test_api_health()
    test_get_all_scholarships()
    test_search_scholarships()
    
    print("üèÅ Tests completed")
````
